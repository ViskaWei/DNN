```python
from torch.utils.data import TensorDataset, DataLoader
```

```python
tx = torch.FloatTensor(train_x).to(device)
ty = torch.FloatTensor(train_y).to(device)
vx = torch.FloatTensor(test_x).to(device)
vy = torch.FloatTensor(test_y).to(device)
```

```python
tt = torch.FloatTensor(test_x[:10]).to(device)
```

```python
train_data = TensorDataset(tx, ty)
test_data = TensorDataset(vx, vy)
```

```python
batch_size = 512
trainloader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
testloader  = DataLoader(test_data, batch_size=batch_size, shuffle=False)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F


class Model(nn.Module):

    def __init__(self):
        super(Model, self).__init__()
        self.fc1 = nn.Linear(20, 32) 
        self.fc2 = nn.Linear(32, 16)
        self.fc3 = nn.Linear(16, 5)
        self.relu = nn.LeakyReLU()
        self.dp = nn.Dropout(p=0.1)
        self.dp2 = nn.Dropout(p=0.05)


    def forward(self, x):
    
        x = torch.flatten(x, 1) # flatten all dimensions except the batch dimension
        x = self.relu(self.fc1(x))
        x = self.dp(x)
        x = self.relu(self.fc2(x))
        x = self.dp2(x)
        x = self.fc3(x)
        return x


model = Model()
model.to(device)
print(model)
```

```python
model(tt)
```

```python
import torch.optim as optim

criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)
```

```python
epochs=100
```

```python
for epoch in range(epochs):  # loop over the dataset multiple times
    for i, data in enumerate(trainloader, 0):
        # get the inputs; data is a list of [inputs, labels]
        inputs, labels = data

        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
    if epoch % 10 == 0:
        print(f"E{epoch}_{loss.item():.03f}", end=" ")

print(f'Finished {loss.item():.03f}')
```

```python
pred= torch.empty_like(vy, device=device)
losses = 0.0
with torch.no_grad():
    for i, data in enumerate(testloader):
        start = i*batch_size
        end = start + batch_size
        images, labels = data
        out = model(images)
        loss = criterion(out, labels)
        losses += loss.item()
        pred[start:end] = out
#         gnd[star:end] = labels
    losses /= pred.size()[0]
print(losses)
```

```python
pred_y = pred.cpu()
```
